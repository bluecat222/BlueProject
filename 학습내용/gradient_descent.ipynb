```python
import numpy as np
```

```python
def prediction(theta_0, theta_1, x):
    """주어진 학습 데이터 벡터 x에 대해서 모든 예측 값을 벡터로 리턴하는 함수"""
    
    return theta_0 + theta_1 * x
    
    
def prediction_difference(theta_0, theta_1, x, y):
    """모든 예측 값들과 목표 변수들의 오차를 벡터로 리턴해주는 함수"""
    
    h = prediction(theta_0, theta_1, x)
    
    return h - y
    
def gradient_descent(theta_0, theta_1, x, y, iterations, alpha):
    """주어진 theta_0, theta_1 변수들을 경사 하강률 하면서 업데이트 해주는 함수"""
    
    m = len(x)
    
    for iter in range(iterations):
        
        sum = 0
        sum2 = 0
        for i in range(m):
            sum += prediction_difference(theta_0, theta_1, x[i], y[i])
            sum2 += prediction_difference(theta_0, theta_1, x[i], y[i]) * x[i]
        
        theta_0 -= alpha * (1/m) * sum
        theta_1 -= alpha * (1/m) * sum2
        
        print(theta_0, theta_1)
        
    return theta_0, theta_1
```

```python
# 입력 변수(집 크기) 초기화 (모든 집 평수 데이터를 1/10 크기를 줄임)
house_size = np.array([0.9, 1.4, 2, 2.1, 2.6, 3.3, 3.35, 3.9, 4.4, 4.7, 5.2, 5.75, 6.7, 6.9])

# 목표 변수(집 크기) 초기화 (모든 집 평수 데이터를 1/10 크기를 줄임)
house_price = np.array([0.3, 0.75, 0.45, 1.1, 1.45, 0.9, 1.8, 0.9, 1.5, 2.2, 1.75, 2.3, 2.49, 2.6])

# theta 값을 초기화 (아무 값이나 시작함)
theta_0 = 2.5
theta_1 = 0

# 학습률 0.1로 200번 경사 하강
theta_0, theta_1 = gradient_descent(theta_0, theta_1, house_size, house_price, 300, 0.1)

theta_0, theta_1
```
